```python
def decision(x):# es un dato
    if x == "A estrenar" or x == "En construcción" or pd.isnull(x):
        return 0
    else:
        return int(x)
```

```python
df["antigüedad"] = df["antigüedad"].apply(decision)
```

```python
def calculo(x):# x es una fila del dataframe
    if x["metroscuadradostotales"] !=0 and x["cuartos"] !=0:
        return [x["precio"] / x["metroscuadradostotales"],x["precio"]/x["cuartos"]]
    else:
        return [0,0]
```

```python
df[["info1","info2"]] = df.apply(calculo, axis = 1, result_type = "expand")
```
