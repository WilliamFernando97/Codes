```python
dfmar = pd.read_csv("../Scripts/data/sales_2017_marzo.csv")
dfmar
```

```python
df = pd.concat([dfene,dffeb,dfmar])
```

```python
df
```

```python
stores = pd.read_csv("../Scripts/data/stores.csv")
stores
```

```python
df = df.merge(stores,left_on = "store_nbr", right_on="store_nbr", how="left")
```

```python
# Cual es la provincia que mas unidades vende.

dfg = df.groupby("state").agg({"unit_sales":np.sum}).reset_index()

dfg.sort_values(by="unit_sales",ascending=False)
```
