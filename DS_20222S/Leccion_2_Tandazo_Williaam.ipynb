> Pregunta 1 
El estado que más vuelos ha recibido en total durante el año 2000, 2001, 2002?

```python
flight.dtypes
flight["Year"] = flight["FLIGHTDATE"].astype(str)
flight["Year"] = flight["Year"].str.slice(0,4)
```

```python
flight1 = flight[["Year","DESTSTATE"]]
flight1 = flight1.groupby(["Year","DESTSTATE"]).size()
flight1 = flight1.reset_index(name="Times").astype(int)

flight1.groupby("Year")["Times"].sum()
flight1
```
